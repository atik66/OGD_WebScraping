```python
import time
import requests
import os
import pandas as pd
import os
import pandas as pd
import csv
import xlrd
import openpyxl
from PyPDF2 import PdfReader
import time
from urllib.parse import unquote,urlparse

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException

driver.get('http://data.gov.bd/dataset')
import logging


datasets_info={}

a_elements = driver.find_elements(By.XPATH, '//div[@class="ctools-collapsible-content"]/div/ul/li/a[@class="facetapi-inactive"]')
k=0
topic=[]

def identify_header(file_path, num_lines=10, sheet_name=0):
    # print("Processing: " + file_path)
    if file_path.endswith('.csv'):
        return identify_csv_header(file_path, num_lines)
    elif file_path.endswith('.xls'):
        return identify_xls_header(file_path, num_lines, sheet_name)
    elif file_path.endswith('.xlsx') or file_path.endswith('.xlb'):
        return identify_excel_header(file_path, num_lines, sheet_name)

    elif file_path.endswith('.pdf'):
        return identify_pdf_header(file_path)
    else:
        raise ValueError("Unsupported file format")

def identify_csv_header(file_path, num_lines=10):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")

    try:
        df_sample_first = pd.read_csv(file_path, nrows=1)  # Read only the first row
        header_first = df_sample_first.columns.tolist()

        df_sample_second = pd.read_csv(file_path, skiprows=1, nrows=1)  # Skip the first row and read the second row
        header_second = df_sample_second.columns.tolist()



        if header_first is not None and len(header_first)>=len(header_second) :
            total_rows = len(pd.read_csv(file_path))-1;
            total_columns = len(header_first)
            empty_cells = df_sample_first.isnull().sum().sum()
            return 0, header_first, total_rows, total_columns, empty_cells

        elif header_second is not None:
            total_rows = len(pd.read_csv(file_path))-2;
            total_columns = len(header_second)
            empty_cells = df_sample_second.isnull().sum().sum()
            return 1, header_second, total_rows, total_columns, empty_cells

        else:
            # print(f"No header found in file: {file_path}")
            return None, None, None, None, None

    except UnicodeDecodeError:
        try:
            df_sample = pd.read_csv(file_path, nrows=num_lines, encoding='latin1')
            header = df_sample.columns.tolist()
            if header:
                total_rows = len(df_sample)
                total_columns = len(header)
                empty_cells = df_sample.isnull().sum().sum()
                return 0, header, total_rows, total_columns, empty_cells
        except Exception as e:
            # print(f"Error reading {file_path}: {e}")
            return None, None, None, None, None

    except Exception as e:
        # print(f"Error reading {file_path}: {e}")
        return None, None, None, None, None


#     header = df_sample.columns.tolist()
#     if header is None:
#         print(f"No header found in file: {file_path}")
#         return None, None, None, None, None

#     total_rows = len(df_sample)
#     total_columns = len(header)
#     empty_cells = df_sample.isnull().sum().sum()

#     return 0, header, total_rows, total_columns, empty_cells

def identify_xls_header(file_path, num_lines=10, sheet_name=0):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")

    try:
        workbook = xlrd.open_workbook(file_path)
        sheet = workbook.sheet_by_index(sheet_name)

        potential_headers = []
        for i in range(min(num_lines, sheet.nrows)):
            row = sheet.row_values(i)
            if all(isinstance(cell, str) for cell in row):
                potential_headers.append(row)

        # Check if the first row or combined first two rows form a header
        if potential_headers:
            if len(potential_headers) == 1:
                header = potential_headers[0]
            elif len(potential_headers) >= 2:
                header = potential_headers[0] + potential_headers[1]
                if all(not cell for cell in header[len(potential_headers[0]):]):
                    header = potential_headers[0]  # Handle case where second row is empty

            total_rows = sheet.nrows - len(potential_headers)
            total_columns = len(header)
            empty_cells = sum(sheet.cell(r, c).value == '' for r in range(len(potential_headers), sheet.nrows) for c in range(sheet.ncols))
            return len(potential_headers) - 1, header, total_rows, total_columns, empty_cells

    except Exception as e:
        #  print(f"Error reading XLS file {file_path}: {e}")
        return None, None, None, None, None

    return None, None, None, None, None

def identify_excel_header(file_path, num_lines=10, sheet_name=0):

    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")

    try:
        df_sample = pd.read_excel(file_path, sheet_name=sheet_name, nrows=num_lines)
        header = df_sample.columns.tolist()
        if header:
            total_rows = len(df_sample)
            total_columns = len(header)
            empty_cells = df_sample.isnull().sum().sum()
            return 0, header, total_rows, total_columns, empty_cells
    except Exception as e:
        # print(f"Error reading Excel file {file_path}: {e}")
        return None, None, None, None, None

    return None, None, None, None, None

def identify_pdf_header(file_path, num_pages=1):
    try:
        with open(file_path, 'rb') as file:
            reader = PdfReader(file)
            text = ''
            for page_num in range(min(num_pages, len(reader.pages))):
                page = reader.pages[page_num]
                text += page.extract_text()

            lines = text.split('\n')
            for i, line in enumerate(lines):
                if line.strip():
                    return i, line.split(','), len(lines), len(line.split(',')), sum(1 for line in lines for cell in line.split(',') if cell.strip() == '')
    except Exception as e:
         print(f"Error reading PDF file {file_path}: {e}")
    return None, None, None, None, None

def detect_time_dimension(header):
    f=0
    time_keywords = ['quarter', 'week', 'day', 'time', 'date', 'month', 'year']
    header_lower = [col.lower() for col in header]
    # time_keywords_lower = [keyword.lower() for keyword in time_keywords]

    # time_columns = [col for col in header_lower if any(keyword in col for keyword in time_keywords_lower)]

    for k in header_lower:
      if 'year' in k.lower():  # Convert k to lowercase for case-insensitive comparison
         f = 1
         break

    if f==0:
     for col in header_lower:
        # print(col)
        if col and len(col) >= 4 and col[0] > '0' and col[0] <= '9':
            header_lower.append('year')
            # print("Detected year: " )
            break

    keyword_indices = {}  # Dictionary to store the index of matching keywords
    for keyword_idx, keyword in enumerate(time_keywords):
      for idx, col in enumerate(header_lower):
         if keyword.lower() in col.lower():
            keyword_indices[keyword] = keyword_idx
            break

    if keyword_indices:
       min_index_keyword = min(keyword_indices, key=keyword_indices.get)
      #  print(min_index_keyword)
       k=time_keywords[keyword_indices[min_index_keyword]]
      #  print(k)
       return k
    else:
       return None



def detect_geographical_dimension(header):
    geo_keywords = ['longitude', 'latitude', 'location', 'union', 'thana', 'city', 'district', 'division', 'province', 'region', 'state', 'country']

    # Convert header column names to lowercase
    header_lower = [col.lower() for col in header]

    # Dictionary to store the first occurrence index of each keyword
    keyword_indices = {}  # Dictionary to store the index of matching keywords
    for keyword_idx, keyword in enumerate(geo_keywords):
      for idx, col in enumerate(header_lower):
         if keyword.lower() in col.lower():
            keyword_indices[keyword] = keyword_idx
            break

    if keyword_indices:
       min_index_keyword = min(keyword_indices, key=keyword_indices.get)
      #  print(min_index_keyword)
       k=geo_keywords[keyword_indices[min_index_keyword]]
      #  print(k)
       return k
    else:
       return None



def columnparse(dataset_name, category_datasets, newfolder):
    columns_info = []

    pos = newfolder.find('/')
    newfilepath = "/content/http:/" + newfolder[pos+2:]

    # print("Searching for CSV files..."+ dataset_name)
    csv_files = [file for file in os.listdir(newfilepath) if file == dataset_name + ".csv" and file.endswith('.csv')]
    if csv_files:
        file = csv_files[0]  # Process the first CSV file
        file_path = os.path.join(newfilepath, file)
        if os.path.isfile(file_path):
            try:
                # print(f"Processing CSV file: {file}")
                header_index, header, total_rows, total_columns, empty_cells = identify_header(file_path)
                if header is not None:
                    time_columns = detect_time_dimension(header)
                    geo_columns = detect_geographical_dimension(header)
                    columns_info.append({
                        'file_name': file,
                        'header_index': header_index,
                        'header': header,
                        'total_rows': total_rows,
                        'total_columns': total_columns,
                        'empty_cells': empty_cells,
                        'time_columns': time_columns,
                        'geo_columns': geo_columns
                    })
                    return columns_info  # Only process one file, so return immediately
                else:
                    print(f"No header found in file: {file_path}")
            except ValueError as e:
                print(e)
            except FileNotFoundError as e:
                print(e)
    else:
        csv_files = [file for file in os.listdir(newfilepath) if file.endswith('.csv')]

        if csv_files:
          file = csv_files[0]  # Process the first CSV file
          file_path = os.path.join(newfilepath, file)
          if os.path.isfile(file_path):
            try:
                # print(f"Processing CSV file: {file}")
                header_index, header, total_rows, total_columns, empty_cells = identify_csv_header(file_path)
                if header is not None:
                    time_columns = detect_time_dimension(header)
                    geo_columns = detect_geographical_dimension(header)
                    columns_info.append({
                        'file_name': file,
                        'header_index': header_index,
                        'header': header,
                        'total_rows': total_rows,
                        'total_columns': total_columns,
                        'empty_cells': empty_cells,
                        'time_columns': time_columns,
                        'geo_columns': geo_columns
                    })
                    return columns_info  # Only process one file, so return immediately
                else:
                    print(f"No header found in file: {file_path}")
            except ValueError as e:
                print(e)
            except FileNotFoundError as e:
                print(e)


    # print("No CSV files found, searching for other file types...")
    # If no CSV files were processed, proceed with other file types
    for file in os.listdir(newfilepath):
        if file.endswith('.csv'):
            continue  # Skip CSV files as they were already processed

        file_path = os.path.join(newfilepath, file)
        if os.path.isfile(file_path):
            try:
                print(f"Processing file: {file}")
                header_index, header, total_rows, total_columns, empty_cells = identify_header(file_path)
                if header is not None:
                    time_columns = detect_time_dimension(header)
                    geo_columns = detect_geographical_dimension(header)
                    columns_info.append({
                        'file_name': file,
                        'header_index': header_index,
                        'header': header,
                        'total_rows': total_rows,
                        'total_columns': total_columns,
                        'empty_cells': empty_cells,
                        'time_columns': time_columns,
                        'geo_columns': geo_columns
                    })
                    return columns_info  # Only process one file, so return immediately
                else:
                    print(f"No header found in file: {file_path}")
            except ValueError as e:
                print(e)
            except FileNotFoundError as e:
                print(e)

    print("No valid files processed.")
    return columns_info




    # file=newfilepath"
    # print("FFFFF"+file)
    # if os.path.exists(file):
    #     df = pd.read_csv(file, nrows=0)  # Read only the header
    #     columns = df.columns.tolist()
    #     print(columns)




def extract_info(url,dataset_name,category_datasets,newfolder):
     source=""
     downloadable=""
     dataset_size_mb = None
     columns={}
     category_datasets={}
     filetype={}
     driver2= web_driver()
     driver2.get(url)
     topic_name= driver2.find_elements(By.XPATH, '//div[@class="content"]/div/div/div/a[@class="name"]')
     description_details = driver2.find_elements(By.XPATH, '//div[@class="field field-name-body field-type-text-with-summary field-label-hidden"]/div/div/p')
     topic = [elem.text for elem in topic_name]
     description = [elem.text for elem in description_details]
     formate =driver2.find_elements(By.XPATH, '//div[@id="data-and-resources"]/div/div/ul/li/div/a/span')
     download=driver2.find_elements(By.XPATH, '//div[@id="data-and-resources"]/div/div/ul/li/div/span[@class="count-download"]')
    #  publisher_element = driver2.find_element(By.XPATH, '//section[@class="field-group-table group_additional"]/table/tbody/tr/td/div/div/div/a')
     try:
      modified_date_element = driver2.find_element(By.XPATH,'//section[@class="field-group-table group_additional"]/table/tbody/tr/td/div[@class="field-name-field-modified-date"]')
     except FileNotFoundError as e:
                print(e)
     release_date_element = driver2.find_element(By.XPATH, '//section[@class="field-group-table group_additional"]/table/tbody/tr/td/div[@class="field-name-field-release-date"]')
     license_element = driver2.find_element(By.XPATH, '//section[@class="field-group-table group_additional"]/table/tbody/tr/td/div[@class="field field-name-field-license field-type-text field-label-hidden"]/div/div')


     graphical_description = 0
     iframe_elements = driver2.find_elements(By.TAG_NAME, 'iframe')
     if iframe_elements:
       graphical_description = 1




     source_elements = driver2.find_elements(By.XPATH, '//div[@class="field field-name-field-additional-info field-type-double-field field-label-above"]/div/div/div/table/tbody/tr/td')
     if source_elements:
            c=0
            for td in source_elements:
                if c == 1:
                    source = td.text.strip()
                    break
                else:
                    c=c+1


     data_links = driver2.find_elements(By.XPATH, '//div[@id="data-and-resources"]/div/div/ul/li/div/span/a[@class="btn btn-primary data-link"]')
     for link in data_links:

           href = link.get_attribute('href')
          #  print(href)
           response = requests.get(href)
           if response.status_code == 200:
        # Extract the filename from the Content-Disposition header if available
                content_disposition = response.headers.get('Content-Disposition')
                if content_disposition:
                     filename = unquote(content_disposition.split('filename=')[-1].strip('"'))
                    #  print("atik"+filename)
                     file=filename
                     pos = file.rfind('/')
                     extensionPos=file.rfind('.')
                     extension=file[extensionPos+1:]
                     filename=file[pos+1:]
                     dataset_size = int(response.headers.get('Content-Length', 0))

                # Convert bytes to megabytes
                     dataset_size_mb = dataset_size / (1024 * 1024)
                     print(f"Dataset Size: {dataset_size_mb:.2f} MB")
                    #  print(filename)
                else:
            # Extract filename from URL if Content-Disposition header is not available
                   parsed_url = urlparse(href)
                   filename = os.path.basename(parsed_url.path)
                   filename = unquote(filename)
                  #  print("atik"+filename)
                   file=filename
                   pos = file.rfind('/')
                   extensionPos=file.rfind('.')
                   extension=file[extensionPos+1:]
                   filename=file[pos+1:]
                  #  print(filename)

                if filename=="":
                   print("file Not dowload able")
                   if extension=="":
                      extension="None"
                   filetype[extension] = {
                         "downloadable":"NO",
                         }
                   continue
        # Handle case where the filename does not contain an extension
                elif '.' not in filename:


            # Attempt to get the extension from the Content-Type header
                  content_type = response.headers.get('Content-Type')
                  if content_type:
                     extension = content_type.split('/')[-1]
                     filename = f"{filename}.{extension}"

                file_path = os.path.join(newfolder, filename)
                # pos=file_path.find('/');
                # newfilepath="/content/http:/"+file_path[pos+2:]

                # print("DD"+newfilepath)
                with open(file_path, 'wb') as f:
                     f.write(response.content)
                    #  print(f"File '{filename}' downloaded successfully and saved in '{file_path}'.")
                     filetype[extension] = {
                         "downloadable":"YES",
                         }

                    #  if extension=='csv':
                    #        print("Reading file:", newfilepath)
                          #  columnparse(newfilepath)

           else:
                  print("Failed to download the file.")
          #  href = link.get_attribute('href')
          # #  print(href)
          #  response = requests.get(href)
          #  if response.status_code == 200:
          #           print("download")
          #           downloadable="Yes"
          #           filename = unquote(href.split('/')[-1])
          #           file_path = os.path.join(newfolder, filename)
          #           with open(file_path, 'wb') as f:
          #               f.write(response.content)
          #               # print(f"File '{filename}' downloaded successfully and saved in '{file_path}'.")
          #  else:
          #    print("Failed to download the file.")

# Get the text content of each element
    #  publisher = publisher_element.text.strip()
     modified_date = modified_date_element.text.strip()
     release_date = release_date_element.text.strip()
     license = license_element.text.strip()




     dataset_formate_name = set()
     dataset_download_count = set()
     for link in formate:
             dataset_formate = link.text.strip()
             dataset_formate_name.add(dataset_formate)
     for link in download:
             download_count = link.text.strip()
             dataset_download_count.add(download_count)


     category_={
            "dataset_Formate": dataset_formate_name,
            "download": dataset_download_count,
            "description": description,
            "graphical_description":graphical_description,
            "category_topic": topic,
            # "publisher":publisher,
            "modified_date":modified_date  ,
            "release_date_element":release_date,
            "license_element":license,
            "source":source,
            "downloadable":filetype,
            "size":dataset_size_mb,
        }
     return category_

# traverse_directories(root_dir,category_datasets)



def navigate_pagination(driver):
    try:
        next_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//li[@class="pager-next"]/a'))
        )
        return next_button
    except TimeoutException:
        print("Next button not found or timeout reached.")
        return None

for a_element in a_elements:
    category_datasets = {}
    if k == 13:
        break

    category = {}
    columns = {}
    category = a_element.text
    category_name=category.split(' (')[0]
    if category_name=="Survey & Census":
       continue
    print(category_name)
    link = a_element.get_attribute('href')
    driver1 = web_driver()
    driver1.get(link)
    folder_path = str(link)
    print(folder_path)
    while True:
        datasets = driver1.find_elements(By.XPATH, '//div[@class="view-content"]/div/article/div/h2/a')
        for dataset in datasets:
            dataset_name = dataset.text
            url = dataset.get_attribute('href')
            folder = str(url[26:])
            print(folder)
            newfolder = folder_path + folder
            print(newfolder)
            os.makedirs(newfolder, exist_ok=True)
            category = extract_info(url, dataset_name, category_datasets, newfolder)
            columns = columnparse(dataset_name, category_datasets, newfolder)
            category_datasets[dataset_name] = {
                "category": category,
                "columns": columns
            }

        next_button = navigate_pagination(driver1)
        if next_button:
            next_button.click()
            # time.sleep(2)
        else:
            break

    datasets_info[category_name] = {  # Store info in the higher-level dictionary with unique key
        "category_name": category_name,
        "number_of_datasets": len(category_datasets),
        "datasets": category_datasets,
    }
    k = k + 1





for category, info in datasets_info.items():
    print(f"Category: {info['category_name']}")
    print(f"Number of Datasets: {info['number_of_datasets']}")
    for dataset, details in info['datasets'].items():
        print(f"  Dataset Name: {dataset}")
        print(f"    Category:")
        for key, value in details['category'].items():
            print(f"      {key}: {value}")
        print(f"    Columns: {details['columns']}")
        print("\n")


```

```python
import time
import requests
import os
import pandas as pd
import os
import pandas as pd
import csv
import xlrd
import openpyxl
from PyPDF2 import PdfReader
import time
from urllib.parse import unquote,urlparse

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException

driver.get('http://data.gov.bd/dataset')
import logging


datasets_info={}

a_elements = driver.find_elements(By.XPATH, '//div[@class="ctools-collapsible-content"]/div/ul/li/a[@class="facetapi-inactive"]')
k=0
topic=[]

def identify_header(file_path, num_lines=10, sheet_name=0):
    # print("Processing: " + file_path)
    if file_path.endswith('.csv'):
        return identify_csv_header(file_path, num_lines)
    elif file_path.endswith('.xls'):
        return identify_xls_header(file_path, num_lines, sheet_name)
    elif file_path.endswith('.xlsx') or file_path.endswith('.xlb'):
        return identify_excel_header(file_path, num_lines, sheet_name)

    elif file_path.endswith('.pdf'):
        return identify_pdf_header(file_path)
    else:
        raise ValueError("Unsupported file format")

def identify_csv_header(file_path, num_lines=10):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")

    try:
        df_sample_first = pd.read_csv(file_path, nrows=1)  # Read only the first row
        header_first = df_sample_first.columns.tolist()

        df_sample_second = pd.read_csv(file_path, skiprows=1, nrows=1)  # Skip the first row and read the second row
        header_second = df_sample_second.columns.tolist()



        if header_first is not None and len(header_first)>=len(header_second) :
            total_rows = len(pd.read_csv(file_path))-1;
            total_columns = len(header_first)
            empty_cells = df_sample_first.isnull().sum().sum()
            return 0, header_first, total_rows, total_columns, empty_cells

        elif header_second is not None:
            total_rows = len(pd.read_csv(file_path))-2;
            total_columns = len(header_second)
            empty_cells = df_sample_second.isnull().sum().sum()
            return 1, header_second, total_rows, total_columns, empty_cells

        else:
            # print(f"No header found in file: {file_path}")
            return None, None, None, None, None

    except UnicodeDecodeError:
        try:
            df_sample = pd.read_csv(file_path, nrows=num_lines, encoding='latin1')
            header = df_sample.columns.tolist()
            if header:
                total_rows = len(df_sample)
                total_columns = len(header)
                empty_cells = df_sample.isnull().sum().sum()
                return 0, header, total_rows, total_columns, empty_cells
        except Exception as e:
            # print(f"Error reading {file_path}: {e}")
            return None, None, None, None, None

    except Exception as e:
        # print(f"Error reading {file_path}: {e}")
        return None, None, None, None, None


#     header = df_sample.columns.tolist()
#     if header is None:
#         print(f"No header found in file: {file_path}")
#         return None, None, None, None, None

#     total_rows = len(df_sample)
#     total_columns = len(header)
#     empty_cells = df_sample.isnull().sum().sum()

#     return 0, header, total_rows, total_columns, empty_cells

def identify_xls_header(file_path, num_lines=10, sheet_name=0):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")

    try:
        workbook = xlrd.open_workbook(file_path)
        sheet = workbook.sheet_by_index(sheet_name)

        potential_headers = []
        for i in range(min(num_lines, sheet.nrows)):
            row = sheet.row_values(i)
            if all(isinstance(cell, str) for cell in row):
                potential_headers.append(row)

        # Check if the first row or combined first two rows form a header
        if potential_headers:
            if len(potential_headers) == 1:
                header = potential_headers[0]
            elif len(potential_headers) >= 2:
                header = potential_headers[0] + potential_headers[1]
                if all(not cell for cell in header[len(potential_headers[0]):]):
                    header = potential_headers[0]  # Handle case where second row is empty

            total_rows = sheet.nrows - len(potential_headers)
            total_columns = len(header)
            empty_cells = sum(sheet.cell(r, c).value == '' for r in range(len(potential_headers), sheet.nrows) for c in range(sheet.ncols))
            return len(potential_headers) - 1, header, total_rows, total_columns, empty_cells

    except Exception as e:
        #  print(f"Error reading XLS file {file_path}: {e}")
        return None, None, None, None, None

    return None, None, None, None, None

def identify_excel_header(file_path, num_lines=10, sheet_name=0):

    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")

    try:
        df_sample = pd.read_excel(file_path, sheet_name=sheet_name, nrows=num_lines)
        header = df_sample.columns.tolist()
        if header:
            total_rows = len(df_sample)
            total_columns = len(header)
            empty_cells = df_sample.isnull().sum().sum()
            return 0, header, total_rows, total_columns, empty_cells
    except Exception as e:
        # print(f"Error reading Excel file {file_path}: {e}")
        return None, None, None, None, None

    return None, None, None, None, None

def identify_pdf_header(file_path, num_pages=1):
    try:
        with open(file_path, 'rb') as file:
            reader = PdfReader(file)
            text = ''
            for page_num in range(min(num_pages, len(reader.pages))):
                page = reader.pages[page_num]
                text += page.extract_text()

            lines = text.split('\n')
            for i, line in enumerate(lines):
                if line.strip():
                    return i, line.split(','), len(lines), len(line.split(',')), sum(1 for line in lines for cell in line.split(',') if cell.strip() == '')
    except Exception as e:
         print(f"Error reading PDF file {file_path}: {e}")
    return None, None, None, None, None

def detect_time_dimension(header):
    f=0
    time_keywords = ['quarter', 'week', 'day', 'time', 'date', 'month', 'year']
    header_lower = [col.lower() for col in header]
    # time_keywords_lower = [keyword.lower() for keyword in time_keywords]

    # time_columns = [col for col in header_lower if any(keyword in col for keyword in time_keywords_lower)]

    for k in header_lower:
      if 'year' in k.lower():  # Convert k to lowercase for case-insensitive comparison
         f = 1
         break

    if f==0:
     for col in header_lower:
        # print(col)
        if col and len(col) >= 4 and col[0] > '0' and col[0] <= '9':
            header_lower.append('year')
            # print("Detected year: " )
            break

    keyword_indices = {}  # Dictionary to store the index of matching keywords
    for keyword_idx, keyword in enumerate(time_keywords):
      for idx, col in enumerate(header_lower):
         if keyword.lower() in col.lower():
            keyword_indices[keyword] = keyword_idx
            break

    if keyword_indices:
       min_index_keyword = min(keyword_indices, key=keyword_indices.get)
      #  print(min_index_keyword)
       k=time_keywords[keyword_indices[min_index_keyword]]
      #  print(k)
       return k
    else:
       return None



def detect_geographical_dimension(header):
    geo_keywords = ['longitude', 'latitude', 'location', 'union', 'thana', 'city', 'district', 'division', 'province', 'region', 'state', 'country']

    # Convert header column names to lowercase
    header_lower = [col.lower() for col in header]

    # Dictionary to store the first occurrence index of each keyword
    keyword_indices = {}  # Dictionary to store the index of matching keywords
    for keyword_idx, keyword in enumerate(geo_keywords):
      for idx, col in enumerate(header_lower):
         if keyword.lower() in col.lower():
            keyword_indices[keyword] = keyword_idx
            break

    if keyword_indices:
       min_index_keyword = min(keyword_indices, key=keyword_indices.get)
      #  print(min_index_keyword)
       k=geo_keywords[keyword_indices[min_index_keyword]]
      #  print(k)
       return k
    else:
       return None



def columnparse(dataset_name, category_datasets, newfolder):
    columns_info = []

    pos = newfolder.find('/')
    newfilepath = "/content/http:/" + newfolder[pos+2:]

    # print("Searching for CSV files..."+ dataset_name)
    csv_files = [file for file in os.listdir(newfilepath) if file == dataset_name + ".csv" and file.endswith('.csv')]
    if csv_files:
        file = csv_files[0]  # Process the first CSV file
        file_path = os.path.join(newfilepath, file)
        if os.path.isfile(file_path):
            try:
                # print(f"Processing CSV file: {file}")
                header_index, header, total_rows, total_columns, empty_cells = identify_header(file_path)
                if header is not None:
                    time_columns = detect_time_dimension(header)
                    geo_columns = detect_geographical_dimension(header)
                    columns_info.append({
                        'file_name': file,
                        'header_index': header_index,
                        'header': header,
                        'total_rows': total_rows,
                        'total_columns': total_columns,
                        'empty_cells': empty_cells,
                        'time_columns': time_columns,
                        'geo_columns': geo_columns
                    })
                    return columns_info  # Only process one file, so return immediately
                else:
                    print(f"No header found in file: {file_path}")
            except ValueError as e:
                print(e)
            except FileNotFoundError as e:
                print(e)
    else:
        csv_files = [file for file in os.listdir(newfilepath) if file.endswith('.csv')]

        if csv_files:
          file = csv_files[0]  # Process the first CSV file
          file_path = os.path.join(newfilepath, file)
          if os.path.isfile(file_path):
            try:
                # print(f"Processing CSV file: {file}")
                header_index, header, total_rows, total_columns, empty_cells = identify_csv_header(file_path)
                if header is not None:
                    time_columns = detect_time_dimension(header)
                    geo_columns = detect_geographical_dimension(header)
                    columns_info.append({
                        'file_name': file,
                        'header_index': header_index,
                        'header': header,
                        'total_rows': total_rows,
                        'total_columns': total_columns,
                        'empty_cells': empty_cells,
                        'time_columns': time_columns,
                        'geo_columns': geo_columns
                    })
                    return columns_info  # Only process one file, so return immediately
                else:
                    print(f"No header found in file: {file_path}")
            except ValueError as e:
                print(e)
            except FileNotFoundError as e:
                print(e)


    # print("No CSV files found, searching for other file types...")
    # If no CSV files were processed, proceed with other file types
    for file in os.listdir(newfilepath):
        if file.endswith('.csv'):
            continue  # Skip CSV files as they were already processed

        file_path = os.path.join(newfilepath, file)
        if os.path.isfile(file_path):
            try:
                print(f"Processing file: {file}")
                header_index, header, total_rows, total_columns, empty_cells = identify_header(file_path)
                if header is not None:
                    time_columns = detect_time_dimension(header)
                    geo_columns = detect_geographical_dimension(header)
                    columns_info.append({
                        'file_name': file,
                        'header_index': header_index,
                        'header': header,
                        'total_rows': total_rows,
                        'total_columns': total_columns,
                        'empty_cells': empty_cells,
                        'time_columns': time_columns,
                        'geo_columns': geo_columns
                    })
                    return columns_info  # Only process one file, so return immediately
                else:
                    print(f"No header found in file: {file_path}")
            except ValueError as e:
                print(e)
            except FileNotFoundError as e:
                print(e)

    print("No valid files processed.")
    return columns_info




    # file=newfilepath"
    # print("FFFFF"+file)
    # if os.path.exists(file):
    #     df = pd.read_csv(file, nrows=0)  # Read only the header
    #     columns = df.columns.tolist()
    #     print(columns)




def extract_info(url,dataset_name,category_datasets,newfolder):
     source=""
     downloadable=""
     columns={}
     category_datasets={}
     filetype={}
     driver2= web_driver()
     driver2.get(url)
     topic_name= driver2.find_elements(By.XPATH, '//div[@class="content"]/div/div/div/a[@class="name"]')
     description_details = driver2.find_elements(By.XPATH, '//div[@class="field field-name-body field-type-text-with-summary field-label-hidden"]/div/div/p')
     topic = [elem.text for elem in topic_name]
     description = [elem.text for elem in description_details]
     formate =driver2.find_elements(By.XPATH, '//div[@id="data-and-resources"]/div/div/ul/li/div/a/span')
     download=driver2.find_elements(By.XPATH, '//div[@id="data-and-resources"]/div/div/ul/li/div/span[@class="count-download"]')
    #  publisher_element = driver2.find_element(By.XPATH, '//section[@class="field-group-table group_additional"]/table/tbody/tr/td/div/div/div/a')
     modified_date_element = driver2.find_element(By.XPATH,'//section[@class="field-group-table group_additional"]/table/tbody/tr/td/div[@class="field-name-field-modified-date"]')
     release_date_element = driver2.find_element(By.XPATH, '//section[@class="field-group-table group_additional"]/table/tbody/tr/td/div[@class="field-name-field-release-date"]')
     license_element = driver2.find_element(By.XPATH, '//section[@class="field-group-table group_additional"]/table/tbody/tr/td/div[@class="field field-name-field-license field-type-text field-label-hidden"]/div/div')


     graphical_description = 0
     iframe_elements = driver2.find_elements(By.TAG_NAME, 'iframe')
     if iframe_elements:
       graphical_description = 1




     source_elements = driver2.find_elements(By.XPATH, '//div[@class="field field-name-field-additional-info field-type-double-field field-label-above"]/div/div/div/table/tbody/tr/td')
     if source_elements:
            c=0
            for td in source_elements:
                if c == 1:
                    source = td.text.strip()
                    break
                else:
                    c=c+1


     data_links = driver2.find_elements(By.XPATH, '//div[@id="data-and-resources"]/div/div/ul/li/div/span/a[@class="btn btn-primary data-link"]')
     for link in data_links:

           href = link.get_attribute('href')
          #  print(href)
           response = requests.get(href)
           if response.status_code == 200:
        # Extract the filename from the Content-Disposition header if available
                content_disposition = response.headers.get('Content-Disposition')
                if content_disposition:
                     filename = unquote(content_disposition.split('filename=')[-1].strip('"'))
                    #  print("atik"+filename)
                     file=filename
                     pos = file.rfind('/')
                     extensionPos=file.rfind('.')
                     extension=file[extensionPos+1:]
                     filename=file[pos+1:]
                    #  print(filename)
                else:
            # Extract filename from URL if Content-Disposition header is not available
                   parsed_url = urlparse(href)
                   filename = os.path.basename(parsed_url.path)
                   filename = unquote(filename)
                  #  print("atik"+filename)
                   file=filename
                   pos = file.rfind('/')
                   extensionPos=file.rfind('.')
                   extension=file[extensionPos+1:]
                   filename=file[pos+1:]
                  #  print(filename)

                if filename=="":
                   print("file Not dowload able")
                   if extension=="":
                      extension="None"
                   filetype[extension] = {
                         "downloadable":"NO",
                         }
                   continue
        # Handle case where the filename does not contain an extension
                elif '.' not in filename:


            # Attempt to get the extension from the Content-Type header
                  content_type = response.headers.get('Content-Type')
                  if content_type:
                     extension = content_type.split('/')[-1]
                     filename = f"{filename}.{extension}"

                file_path = os.path.join(newfolder, filename)
                # pos=file_path.find('/');
                # newfilepath="/content/http:/"+file_path[pos+2:]

                # print("DD"+newfilepath)
                with open(file_path, 'wb') as f:
                     f.write(response.content)
                    #  print(f"File '{filename}' downloaded successfully and saved in '{file_path}'.")
                     filetype[extension] = {
                         "downloadable":"YES",
                         }

                    #  if extension=='csv':
                    #        print("Reading file:", newfilepath)
                          #  columnparse(newfilepath)

           else:
                  print("Failed to download the file.")
          #  href = link.get_attribute('href')
          # #  print(href)
          #  response = requests.get(href)
          #  if response.status_code == 200:
          #           print("download")
          #           downloadable="Yes"
          #           filename = unquote(href.split('/')[-1])
          #           file_path = os.path.join(newfolder, filename)
          #           with open(file_path, 'wb') as f:
          #               f.write(response.content)
          #               # print(f"File '{filename}' downloaded successfully and saved in '{file_path}'.")
          #  else:
          #    print("Failed to download the file.")

# Get the text content of each element
    #  publisher = publisher_element.text.strip()
     modified_date = modified_date_element.text.strip()
     release_date = release_date_element.text.strip()
     license = license_element.text.strip()




     dataset_formate_name = set()
     dataset_download_count = set()
     for link in formate:
             dataset_formate = link.text.strip()
             dataset_formate_name.add(dataset_formate)
     for link in download:
             download_count = link.text.strip()
             dataset_download_count.add(download_count)


     category_={
            "dataset_Formate": dataset_formate_name,
            "download": dataset_download_count,
            "description": description,
            "graphical_description":graphical_description,
            "category_topic": topic,
            # "publisher":publisher,
            "modified_date":modified_date  ,
            "release_date_element":release_date,
            "license_element":license,
            "source":source,
            "downloadable":filetype,
        }
     return category_

# traverse_directories(root_dir,category_datasets)



def navigate_pagination(driver):
    try:
        next_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//li[@class="pager-next"]/a'))
        )
        return next_button
    except TimeoutException:
        print("Next button not found or timeout reached.")
        return None

for a_element in a_elements:
    category_datasets = {}
    if k == 13:
        break

    category = {}
    columns = {}
    category = a_element.text
    category_name=category.split(' (')[0]
    if category_name=="Survey & Census":
       continue
    print(category_name)
    link = a_element.get_attribute('href')
    driver1 = web_driver()
    driver1.get(link)
    folder_path = str(link)

    while True:
        datasets = driver1.find_elements(By.XPATH, '//div[@class="view-content"]/div/article/div/h2/a')
        for dataset in datasets:
            dataset_name = dataset.text
            url = dataset.get_attribute('href')
            folder = str(url[26:])
            newfolder = folder_path + folder
            os.makedirs(newfolder, exist_ok=True)
            category = extract_info(url, dataset_name, category_datasets, newfolder)
            columns = columnparse(dataset_name, category_datasets, newfolder)
            category_datasets[dataset_name] = {
                "category": category,
                "columns": columns
            }

        next_button = navigate_pagination(driver1)
        if next_button:
            next_button.click()
            # time.sleep(2)
        else:
            break

    datasets_info[category_name] = {  # Store info in the higher-level dictionary with unique key
        "category_name": category_name,
        "number_of_datasets": len(category_datasets),
        "datasets": category_datasets,
    }
    k = k + 1












for category, info in datasets_info.items():
    print(f"Category: {info['category_name']}")
    print(f"Number of Datasets: {info['number_of_datasets']}")
    for dataset, details in info['datasets'].items():
        print(f"  Dataset Name: {dataset}")
        print(f"    Category:")
        for key, value in details['category'].items():
            print(f"      {key}: {value}")
        print(f"    Columns: {details['columns']}")
        print("\n")


```
